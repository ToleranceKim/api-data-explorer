```python
import pandas as pd

#df_stats = pd.read_csv(r'C:\Users\USER\Desktop\my_git\api-data-explorer\analysis\performance_stats_merged.csv')
df_list = pd.read_csv(r'C:\Users\USER\Desktop\my_git\api-data-explorer\analysis\performance_list_merged.csv')
df_box = pd.read_csv(r'C:\Users\USER\Desktop\my_git\api-data-explorer\analysis\boxstats_merged.csv')
print(df_list, df_box)
```

```python
df_box['날짜'] = pd.to_datetime(df_box['날짜'])
df_box['공연시작일'] = pd.to_datetime(df_list['공연시작일'])
df_list['공연종료일'] = pd.to_datetime(df_list['공연종료일'])

def expand_days(row):
    if pd.isnull(row['공연시작일']) or pd.isnull(row['공연종료일']):
        return []
    return pd.date_range(row['공연시작일'], row['공연종료일'], freq='D')

df_list['상연일자리스트'] = df_list.apply(expand_days, axis=1)
df_explode = df_list.explode('상연일자리스트')
df_explode.rename(columns={'상연일자리스트':'상연일'}, inplace=True)
```

```python
# df_box와 상연일 기준으로 merge
df_merged = pd.merge(
    df_explode,
    df_box,
    left_on='상연일',
    right_on='날짜',
    how='inner'
)
```

```python
# 일자별 매출을 공연 수로 균등 분배(가정)
# df_merged 안에 그 날짜에 해당하는 공연이 총 몇 개인지 groupby -> transform
df_merged['공연수_그날'] = df_merged.groupby('상연일')['공연ID'].transform('count')

# 예: 
```
